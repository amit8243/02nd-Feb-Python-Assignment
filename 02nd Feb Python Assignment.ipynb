```python
#Python Assignment
"""Q1. Explain with an example each when to use a for loop and a while loop."""
Ans: 
For Loop: A for loop is used when you know the exact number of times you want to loop through a block of code. 
For example, if you wanted to print out the numbers 1-10, you could use a for loop like this:

for i in range(10):
    print(i)

While Loop: A while loop is used when you do not know the exact number of times you want to loop through a block of code.
For example, if you wanted to keep asking a user for input until they enter the word "quit", you could use a while loop 
like this:

user_input = ""
while user_input != "quit":
    user_input = input("Please enter something: ")
    print("You entered: " + user_input)

print("Goodbye!")

"""Q2. Write a python program to print the sum and product of the first 10 natural numbers using for
and while loop."""
Ans: 
#Using for loop
sum = 0
product = 1
for i in range(1,10):
    sum += i
    product *= i
print("Sum of first 10 natural numbers:",sum)
print("Product of first 10 natural numbers:",product)

#Output
Sum of first 10 natural numbers: 55
Product of first 10 natural numbers: 3628800

#Using while loop
sum = 0
product = 1
i = 1
while i<=10:
    sum += i
    product *= i
    i += 1
print("Sum of first 10 natural numbers:",sum)
print("Product of first 10 natural numbers:",product)

#Output
Sum of first 10 natural numbers: 55
Product of first 10 natural numbers: 3628800

"""Q3. Create a python program to compute the electricity bill for a household.
The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per unit, for 
the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will be charged Rs. 10
per unit, After 300 units and above the user will be charged Rs. 20 per unit. You are required to take the units of 
electricity consumed in a month from the user as input. Your program must pass this test case: when the unit of 
electricity consumed by the user in a month is 310, the total electricity bill should be 2250."""

# Program to calculate electricity bill for a given unit 
unit = float(input("Enter total units consumed: ")) 
  
if unit <= 100: 
    amount = unit * 4.5 
elif unit <= 200: 
    amount = 100 * 4.5 + (unit-100) * 6
elif unit <= 300: 
    amount = 100 * 4.5 + 100 * 6 + (unit-200) * 10
else: 
    amount = 100 * 4.5 + 100 * 6 + 100 * 10 + (unit-300) * 20
  
 print("Total bill = Rs.", amount) 
  
# Test case 1 
unit = 310 
if unit <= 100: 
    amount = unit * 4.5 
elif unit <= 200: 
    amount = 100 * 4.5 + (unit-100) * 6
elif unit <= 300: 
    amount = 100 * 4.5 + 100 * 6 + (unit-200) * 10
else: 
    amount = 100 * 4.5 + 100 * 6 + 100 * 10 + (unit-300) * 20
  
 print("Total bill = Rs.", amount)
# Output: Total bill = Rs. 2250.0

"""Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
that list."""
Ans:
# Using for loop

cube_list = []
for num in range(1,101):
    cube = num**3
    if cube % 4 == 0 or cube % 5 == 0:
        cube_list.append(num)
print(cube_list)

# Using while loop

cube_list = []
num = 1
while num <= 100:
    cube = num**3
    if cube % 4 == 0 or cube % 5 == 0:
        cube_list.append(num)
    num += 1
print(cube_list)

# Output
[1, 4, 5, 8, 9, 16, 17, 20, 25, 27, 32, 33, 36, 40, 45, 49, 50, 64, 65, 68, 72, 81]

"""Q5. Write a program to filter count vowels in the below-given string. string = "I want to become a data scientist"""
Ans: 
    
string = "I want to become a data scientist"

vowels = 0
for char in string: 
    if char in 'aeiouAEIOU': 
        vowels = vowels + 1
print("Number of vowels in the given string:", vowels)

# Output: Number of vowels in the given string: 12
```
